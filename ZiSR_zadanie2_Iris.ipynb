```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()

X = iris.data
y = iris.target
feature_names = iris.feature_names
class_names = iris.target_names

print("Cechy:", feature_names)
print("Klasy:", class_names)
print("Rozmiar danych:", X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```

```python
stats = {}

for class_id in np.unique(y_train):
    class_data = X_train[y_train == class_id]

    stats[class_id] = []

    for i in range(class_data.shape[1]):
        feature = class_data[:, i]

        mean = np.mean(feature)
        median = np.median(feature)
        std = np.std(feature)
        min_val = np.min(feature)
        max_val = np.max(feature)

        stats[class_id].append({
            "mean": mean,
            "median": median,
            "std": std,
            "min": min_val,
            "max": max_val
        })

stats
```

```python
def gaussian_membership(x, mean, std):
    if std == 0:
        std = 1e-6
    return np.exp(-((x - mean) ** 2) / (2 * std ** 2))
```

```python
def classify(sample, stats):

    scores = {}

    for class_id in stats:
        score = 0

        for i, value in enumerate(sample):

            mean = stats[class_id][i]["mean"]
            std = stats[class_id][i]["std"]

            score += gaussian_membership(value, mean, std)

        scores[class_id] = score

    return max(scores, key=scores.get)
```

```python
predictions = []

for sample in X_test:
    pred = classify(sample, stats)
    predictions.append(pred)

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
```

```python
sample = X_test[0]

print("Próbka:", sample)

predicted_class = classify(sample, stats)

print("Przewidziana klasa:", class_names[predicted_class])
print("Prawdziwa klasa:", class_names[y_test[0]])
```
